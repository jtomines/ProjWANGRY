```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import re

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')


# File to Load
H2ODataFile = "../1_Input/number-with-without-access-to-improved-water.csv"

# Load to dataframe
H2O_World_Access_data = pd.read_csv(H2ODataFile)

# Simplify column headers
H2O_World_Access_data = H2O_World_Access_data.rename(columns = {"Population without access to improved water source (people)" : "Population without Access",
                                        "Population with access to improved water source (people)" : "Population with Access"
                                       })


H2O_World_Access_data["% Access"] = H2O_World_Access_data["Population with Access"] * 100 / (H2O_World_Access_data["Population with Access"] + H2O_World_Access_data["Population without Access"])

H2O_World_Access_data
```

```python
# File to Load
H2ODataFile = "../1_Input/water-withdrawals-per-capita.csv"

# Load to dataframe
H2O_World_Withdrawal_data = pd.read_csv(H2ODataFile)

# Simplify column headers
H2O_World_Withdrawal_data
```

```python
H2O_World_Withdrawal_grp = H2O_World_Withdrawal_data.groupby(by="Entity")
H2O_World_Withdrawal_latestYear = pd.DataFrame(H2O_World_Withdrawal_grp["Year"].max())
H2O_World_Withdrawal_latestYear = H2O_World_Withdrawal_latestYear.reset_index()

#H2O_World_Withdrawal_latestYear

H2O_World_Withdrawal_data.head() 
```

```python
Latest_World_Withdrawal = pd.merge(H2O_World_Withdrawal_data, H2O_World_Withdrawal_latestYear, how='inner', on=['Entity', 'Year'])
Latest_World_Withdrawal.head()
```

```python
Latest_World_Water_data = pd.merge(Latest_World_Withdrawal, H2O_World_Access_data, how='left', on=['Entity', 'Year'])

Latest_World_Water_data = Latest_World_Water_data.drop(columns=['Code_y']).rename(columns={'Code_x': 'Code'})

Latest_World_Water_data
```
