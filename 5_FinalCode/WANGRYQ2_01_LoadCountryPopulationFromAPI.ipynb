```python
# Dependencies
import pandas as pd
import requests

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')
```

```python
# setup query url
query_url = 'http://api.population.io:80/1.0/countries'

# setup country list
country=[]

# get list of countries from API
country_response = requests.get(query_url).json()

# remove all non-country items
for item in range(len(country_response['countries'])):
    if country_response['countries'][item] not in ['AFRICA',
                                                   'ASIA',
                                                   'EUROPE',
                                                   'LATIN AMERICA AND THE CARIBBEAN',
                                                   'NORTHERN AMERICA', 'OCEANIA',
                                                   'South America',
                                                   'Other non-specified areas',
                                                   'World',
                                                   'Less developed regions',
                                                   'Less developed regions, excluding China',
                                                   'Less developed regions, excluding least developed countries',
                                                   'Australia/New Zealand',
                                                   'Least developed countries',
                                                   'More developed regions',
                                                   'Sub-Saharan Africa',
                                                   'Western Africa',
                                                   'Western Asia',
                                                   'Western Europe',
                                                   'Western Sahara',
                                                   'South-Central Asia',
                                                   'South-Eastern Asia',
                                                   'Southern Africa',
                                                   'Southern Asia',
                                                   'Southern Europe',
                                                   'Eastern Africa',
                                                   'Eastern Asia',
                                                   'Eastern Europe',
                                                   'Northern Africa',
                                                   'Northern Europe',
                                                   'Middle Africa'
                                                  ]:
        country.append(country_response['countries'][item])

country
```

```python
# Get population for each country

# Setup population list
country_population=[]

# Loop to get population for each country
for country_name in country:
    pop_url = 'http://api.population.io:80/1.0/population/' + country_name + '/2016-01-01/'
    population_response = requests.get(pop_url).json()
    population = population_response['total_population']['population']
    country_population.append([country_name, population])

# Create Data Frame    
population_df = pd.DataFrame(country_population)
population_df = population_df.rename(columns = {0: 'Country', 1: 'Population'})

population_df.head()
```

```python
# Save the data:
population_df.to_csv('../1_Input/DF_Interim/population_data.csv', index=False, encoding='utf-8')
```
