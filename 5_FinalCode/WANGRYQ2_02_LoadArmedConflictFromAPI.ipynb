```python
# Dependencies
import pandas as pd
import requests

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')
```

```python
# Data to store from the json
conflict_id=[]
year=[]
latitude=[]
longitude=[]
country=[]
country_id=[]
best=[]

# Loop through each page
for page in range(99):

    # Build query URL for each page
    query_url = "http://ucdpapi.pcr.uu.se/api/gedevents/18.1?pagesize=1000&page=" + str(page) + "&StartDate=2000-01-01"

    # get page data from API
    conflict_response = requests.get(query_url)
    conflict_json = conflict_response.json()
 
    if page == 99:
        recAmt = 173
    else:
        recAmt = 1000

    # Loop through each conflict item on the page
    for conflict in range(recAmt):
        conflict_id.append(conflict_json["Result"][conflict]["id"])
        year.append(conflict_json["Result"][conflict]["year"])
        latitude.append(conflict_json["Result"][conflict]["latitude"])
        longitude.append(conflict_json["Result"][conflict]["longitude"])
        country.append(conflict_json["Result"][conflict]["country"])
        country_id.append(conflict_json["Result"][conflict]["country_id"])
        best.append(conflict_json["Result"][conflict]["best"])
```

```python
# Create dataframe to hold conflict data
conflict_data = {"Conflict Id": conflict_id,
                 "Year": year,
                 "Latitude": latitude,
                 "Longitude": longitude,
                 "Country": country,
                 "Country Id": country_id,
                 "Best Death Estimate": best
                }
conflict_df = pd.DataFrame(conflict_data)

# Clean up country names to match other dataframes
conflict_df.loc[conflict_df["Country"] == "Russia (Soviet Union)", "Country"] = "Russian Federation"
conflict_df.loc[conflict_df["Country"] == "DR Congo (Zaire)", "Country"] = "Congo"
conflict_df.loc[conflict_df["Country"] == "Macedonia, FYR", "Country"] = "FYR Macedonia"
conflict_df.loc[conflict_df["Country"] == "Cambodia (Kampuchea)", "Country"] = "Cambodia"
conflict_df.loc[conflict_df["Country"] == "Zimbabwe (Rhodesia)", "Country"] = "Zimbabwe"
conflict_df.loc[conflict_df["Country"] == "Serbia (Yugoslavia)", "Country"] = "Serbia"
conflict_df.loc[conflict_df["Country"] == "Madagascar (Malagasy)", "Country"] = "Madagascar"
conflict_df.loc[conflict_df["Country"] == "Yemen (North Yemen)", "Country"] = "Rep of Yemen"
conflict_df.loc[conflict_df["Country"] == "Myanmar (Burma)", "Country"] = "Myanmar"
conflict_df.loc[conflict_df["Country"] == "Ivory Coast", "Country"] = "Cote-d-Ivoire"
conflict_df.loc[conflict_df["Country"] == "Iran", "Country"] = "Islamic Republic of Iran"
conflict_df.loc[conflict_df["Country"] == "Laos", "Country"] = "Lao PDR"
conflict_df.loc[conflict_df["Country"] == "Venezuela", "Country"] = "RB-de-Venezuela"
conflict_df.loc[conflict_df["Country"] == "United States of America", "Country"] = "United States"

conflict_df.head()
```
