```python
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load Water Access File
wateraccess_data_file = "../1_Input/DF_Interim/WorldWaterAccessMean_data.csv"

# Load file
wateraccess_df = pd.read_csv(wateraccess_data_file)

wateraccess_df.head()
```

```python
# Load Water Withdrawal File
waterwithdrawal_data_file = "../1_Input/DF_Interim/WorldWaterWithdrawalMean_data.csv"

# Load file
waterwithdrawal_df = pd.read_csv(waterwithdrawal_data_file)

waterwithdrawal_df.head()
```

```python
# Load Conflict File
conflict_data_file = "../1_Input/DF_Interim/conflict_perCountry.csv"

# Load file
conflict_df = pd.read_csv(conflict_data_file)

conflict_df.head()
```

```python
# Load Homocide File
homocide_data_file = "../1_Input/DF_Interim/homocide_perCountry.csv"

# Load file
homocide_df = pd.read_csv(homocide_data_file)

homocide_df.head()
```

```python
# Merge Water Data Frames
water_df = pd.merge(wateraccess_df, waterwithdrawal_df, how="outer", on="Country")

water_df.head()
```

```python
# Merge Conflict Data Frames
violence_df = pd.merge(conflict_df, homocide_df, how="outer", on="Country")

violence_df.head()
```

```python
# Merge Remaining Data Frames
water_vs_violence = pd.merge(water_df, violence_df, how="outer", on="Country")

water_vs_violence.head()
```

```python
# Save the data:
water_vs_violence.to_csv('../1_Input/DF_Interim/World_Water_vs_Violence.csv', index=False, encoding='utf-8')
```
