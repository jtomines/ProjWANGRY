```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import re

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')


# File to Load
H2ODataFile = "../1_Input/water-withdrawals-per-capita.csv"

# Load to dataframe
H2O_World_Withdrawal_data = pd.read_csv(H2ODataFile)

# Rename columns
H2O_World_Withdrawal_data = H2O_World_Withdrawal_data.rename(columns = {"Entity": "Country"})

# Clean up country names to match other dataframes
H2O_World_Withdrawal_data.loc[H2O_World_Withdrawal_data["Country"] == "Cote d'Ivoire", "Country"] = "Cote-d-Ivoire"
H2O_World_Withdrawal_data.loc[H2O_World_Withdrawal_data["Country"] == "Democratic Republic of Congo", "Country"] = "Congo"
H2O_World_Withdrawal_data.loc[H2O_World_Withdrawal_data["Country"] == "Saint Lucia", "Country"] = "St-Lucia"
H2O_World_Withdrawal_data.loc[H2O_World_Withdrawal_data["Country"] == "Saint Vincent and the Grenadines", "Country"] = "St-Vincent and the Grenadines"

# Display
H2O_World_Withdrawal_data.head()
```
